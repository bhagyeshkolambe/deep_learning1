```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt

# Generate quadratic-like data
torch.manual_seed(42)
x = torch.randn(100, 1)  # Input data
y = x**2 + 2*x + 3 + 0.1 * torch.randn_like(x)  # Quadratic-like function with noise

# Define the model with a non-linear activation function (e.g., ReLU)
class QuadraticModel(nn.Module):
    def __init__(self):
        super(QuadraticModel, self).__init__()
        self.hidden = nn.Linear(1, 5)  # Add a hidden layer
        self.relu = nn.ReLU()
        self.output = nn.Linear(5, 1)

    def forward(self, x):
        x = self.relu(self.hidden(x))
        x = self.output(x)
        return x

# Instantiate the model
model = QuadraticModel()

# Define the loss function and optimizer
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

# Training the model
num_epochs = 1000
for epoch in range(num_epochs):
    # Forward pass
    y_pred = model(x)

    # Compute the loss
    loss = criterion(y_pred, y)

    # Backward pass and optimization
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 100 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')

# Plot the results
with torch.no_grad():
    x_range = torch.linspace(float(min(x.numpy())), float(max(x.numpy())), 100).unsqueeze(1)
    plt.scatter(x, y, label='True data')
    plt.plot(x_range, model(x_range).numpy(), color='red', label='Fitted model')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.show()
```
